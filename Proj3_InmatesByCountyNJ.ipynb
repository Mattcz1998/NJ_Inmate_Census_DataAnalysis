```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('NJDOC_Inmate_Census.csv')
df = pd.DataFrame(data)
```

Total number of inmates in census data:

```python
print("There are", len(df), "inmates documented in this inmate census (as of January 1st, 2021).")
```

Original Bar Chart (provided on https://data.nj.gov/Public-Safety/Inmates-by-County/4z56-bsux):

```python
# Bar Graph (Total Inmates)
plt.figure(figsize=(30,8))
count = df['County'].value_counts().plot(kind = 'bar')

# Titles and labels
plt.xticks(rotation = 90)
plt.title('Inmates by County')
plt.ylabel('Number of Inmates')
plt.xlabel('County')

plt.show()
```

What is the gender distribution of inmates in New Jersey?

```python
plt.figure(figsize=(16,2))
count =  df['Gender'].value_counts()
print(count)

# Horizontal Bar Graph
sns.barplot(count.values,count.index,palette = 'BuGn')
for x,y in enumerate(count.values):
    plt.text(1,x,y,color='k',fontsize=12)
    
plt.title('Gender Distribution')
plt.xlabel('Count')
plt.show()
```

Total MALE inmates by county:

```python
# Bar Graph (Total MALE Inmates)
plt.figure(figsize=(30,8))
ok = df.groupby(["County", "Gender"]).sum()
count = df[df["Gender"] == "M"]['County'].value_counts().plot(kind = 'bar')

# Titles and labels
plt.xticks(rotation = 90)
plt.title('Inmates by County')
plt.ylabel('Number of Inmates')
plt.xlabel('County')

plt.show()
```

Total FEMALE inmates by county:

```python
# Bar Graph (Total FEMALE Inmates)
plt.figure(figsize=(30,8))
count = df[df["Gender"] == "F"]['County'].value_counts().plot(kind = 'bar')

# Titles and labels
plt.xticks(rotation = 90)
plt.title('Inmates by County')
plt.ylabel('Number of Inmates')
plt.xlabel('County')

plt.show()
```

What is the distribution of TERMS AT ADMISSION for all inmates in New Jersey?

```python
plt.figure(figsize=(16,2))
count =  df['Term at Admission (Category)'].value_counts()
print(count)

# Horizontal Bar Graph
sns.barplot(count.values,count.index,palette = 'RdGy')
for x,y in enumerate(count.values):
    plt.text(1,x,y,color='k',fontsize=12)
    
plt.title('Term at Admission Distribution')
plt.xlabel('Count')
plt.show()
```

What is the distribution of LIFE sentences for inmates by county?

```python
# Bar Graph
plt.figure(figsize=(30,8))
ok = df.groupby(["County", "Total Max Term"]).sum()
count = df[df["Total Max Term"] == "LIFE"]['County'].value_counts()


sns.barplot(count.values,count.index,palette = 'RdGy')

# Titles and labels
plt.xticks(rotation = 90)
plt.title('Inmates by County')
plt.ylabel('Number of Inmates')
plt.xlabel('County')

plt.show()
```

What is the total number of inmates in regards to age distribution?

```python
plt.figure(figsize=(50,5))
bins= [18,21,31,41,51,61,71,81,91]
labels = ['0-18','21-30','31-40','41-50','51-60',"61-70", "71-80", "81-90"]
count = pd.cut(df['Age'], bins=bins, labels=labels, right=False).value_counts()
print(count)

# Horizontal Bar Graph
sns.barplot(count.values,count.index,palette = 'RdGy')
for x,y in enumerate(count.values):
    plt.text(1,x,y,color='k',fontsize=12)
    
plt.title('Age Distribution')
plt.xlabel('Count')
plt.show()

# Gotta fix this. Gonna be a horizontal bar graph thats fairly wide. 
# Find a way to insert a RANGE of ages (ex. 0-18, 21-30, 31-40, 41-50, 51-60, 61-70, 71-80, 81-90, 91+)
```

What is the ethnicity distribution of inmates in New Jersey?
